```python
def spherical_coords(num_pts, radius):
    """Return list of uniformly distributed points in a sphere."""
    position_list = []   
    for _ in range(num_pts):
        coords =np.random.normal(0, 1, 3)
        coords *= radius
        position_list.append(list(coords))
    return position_list  
```
